## Hvorfor *Seaborn*?  
[Seaborn](https://seaborn.pydata.org/) er et bibliotek som forenkler prosessen med å lage visualiseringer som linjediagrammer, histogrammer, korrelasjonsmatriser, boksplotter, og mye mer. Seaborn har flere fordeler:  
- Lett å bruke med Pandas DataFrames  
- Har innebygde temaer og farger for vakre visualiseringer  
- Støtter statistisk visualisering, som regressjonslinjer og distribusjonsplotter  
- Gir enklere tilgang til komplekse plott som for eksempel varmekart og pairplots  

I dette seminaret skal vi bruke Seaborn for å visualisere data og trekke innsikt fra grafene.
